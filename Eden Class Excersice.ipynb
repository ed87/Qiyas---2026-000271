```python
# exercise 18 Missing Values Detection
data_with_nan = {
    'A': [1, np.nan, 3],
    'B': [4, np.nan, np.nan],
    'C': [7, 8, np.nan]}
df_nan = pd.DataFrame(data_with_nan)
missing_values_count = df_nan.isnull().sum()
total_missing_values = df_nan.isnull().sum().sum()
print("Missing values in each column:\n", missing_values_count)
print("Total missing values in the dataset:", total_missing_values)
```

```python
#exercise 19 Handling Missing Values
mean_filled_df = df_nan.fillna(df_nan.mean())
median_filled_df = df_nan.fillna(df_nan.median())
print("Mean filled DataFrame:\n", mean_filled_df)
print("Median filled DataFrame:\n", median_filled_df)
```

```python
#exercise 20 Duplicate Records
data_with_duplicates = {
    'Name': ['Abebe', 'Bini', 'sara', 'beti'],
    'Score': [85, 70, 85, 90]}
df_duplicates = pd.DataFrame(data_with_duplicates)
duplicates_identified = df_duplicates.duplicated()
df_no_duplicates = df_duplicates.drop_duplicates()
print("Identified duplicates:\n", duplicates_identified)
print("DataFrame after removing duplicates:\n", df_no_duplicates)
```

```python
#exercise 23 Cleaning a Customer Dataset 
data_customer = { 'Name': ['Alemu', 'Bana', 'Alex', 'dani'],
    'Age': [25, np.nan, -1, -5],
    'City': ['Addis Ababa', 'Gonder', 'Asmara', 'Gonder']}
customer_df = pd.DataFrame(data_customer)
#handling duplicates 
customer_df.drop_duplicates()
#Handling missing values 
customer_df['Age'].fillna(customer_df['Age'].mean())
#standardizing text data 
customer_df['City'] = customer_df['City'].str.title()
print("Cleaned customer dataset:\n", customer_df)
```

```python
#exercise 24: Outlier Detection Using IQR 
sales_data = np.array([100, 150, 200, 250, 300, 1000])  
Q1 = np.percentile(sales_data, 25)
Q3 = np.percentile(sales_data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
outliers = sales_data[(sales_data < lower_bound) | (sales_data > upper_bound)]
print("Outlier values:", outliers)
```

```python
#excercise 25
import pandas as pd
import numpy as np
data_pipeline = {'value': [10, np.nan, -10, -100, -5, np.nan, -20]}
pipeline_df = pd.DataFrame(data_pipeline)

missing_values_count_pipeline = pipeline_df.isnull().sum()
print("Missing values before fill:\n", missing_values_count_pipeline)
pipeline_df['value'] = pipeline_df['value'].fillna(pipeline_df['value'].mean())
pipeline_df.loc[pipeline_df['value'] < 0, 'value'] = np.nan
print("\nFinal DataFrame:\n", pipeline_df)
```
