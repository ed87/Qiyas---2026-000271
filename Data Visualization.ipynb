```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [10, 15, 7, 12]

plt.plot(months, sales, color='green', linestyle='--', marker='s', markersize=8, label='sales')

plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Exercise 2 Data
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]

# Creating the line chart with specifications
plt.plot(
    subjects, 
    scores, 
    color='red', 
    marker='*', 
    linewidth=3, 
    markerfacecolor='yellow', 
    markersize=12,  # Adjusted size so the star and its yellow face color are clearly visible
    label='Scores'
)

# Adding title and legend
plt.title('Exam Scores')
plt.legend()

# Saving the plot
plt.savefig('exam_scores.png', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

# Data
years = [2021, 2022, 2023, 2024]
revenue = [500, 700, 850, 1000]

# Creating the line chart with specifications
plt.plot(
    years, 
    revenue, 
    color='blue', 
    linestyle=':',    # Dotted line
    marker='d',       # Diamond marker
    markersize=8,
    label='Revenue'
)

# Adding labels and title
plt.title('Company Revenue (2021 - 2024)')
plt.xlabel('Year')
plt.ylabel('Revenue')

plt.xticks(years)

# Displaying the legend
plt.legend()

# Saving the plot
plt.savefig('company_revenue.png', bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
departments = ('CS', 'IS', 'SE', 'IT')
students = {
    'Male': np.array([70, 65, 80, 50]),
    'Female': np.array([60, 55, 70, 45])
}

# Using fig, ax = plt.subplots()
fig, ax = plt.subplots()

# 1. Plot the base layer (Male)
male_bars = ax.bar(departments, students['Male'], width=0.7, label='Male')

# 2. Plot the stacked layer (Female) using the 'bottom' parameter
female_bars = ax.bar(departments, students['Female'], width=0.7, bottom=students['Male'], label='Female')

# Add labels inside each section using ax.bar_label()
ax.bar_label(male_bars, label_type='center')
ax.bar_label(female_bars, label_type='center')

# Customizing the chart
ax.set_title('Students by Department and Gender')
ax.set_ylabel('Number of Students')
ax.legend()

# Saving the plot
fig.savefig('students_stacked_bar.png', bbox_inches='tight')
```
